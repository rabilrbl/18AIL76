```python
import numpy as np

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn import datasets, metrics, preprocessing

# load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# set colormaps
colormap = np.array(['red', 'lime', 'black'])

# Fit all
kmeans_labels = KMeans(n_clusters=3).fit(X).labels_
gmm_labels = GaussianMixture(n_components=3).fit(preprocessing.StandardScaler().fit_transform(X)).predict(X)

import matplotlib.pyplot as plt

# set figsize
plt.figure(figsize=(14, 7))

# plot
for i, (title, labels) in  enumerate(zip(['Real', 'KM', 'GMM'], [y, kmeans_labels, gmm_labels])):
    plt.subplot(1,3, i+1)
    plt.scatter(X[:,2], X[:,3], c=colormap[labels], s=40)
    plt.xlabel('Petal length')
    plt.ylabel('Petal width')
    plt.title(title)


# print accuracy
print(f"Accuracy KM: {metrics.accuracy_score(y, kmeans_labels)}")
print(f"Accuracy GMM: {metrics.accuracy_score(y, gmm_labels)}")

plt.show()
```
