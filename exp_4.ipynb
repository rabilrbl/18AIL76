```python
import numpy as np

# define data, x and y
data = np.array(([2, 9], [1,5], [3, 6]), dtype=float)
X = data / np.amax(data, axis=0)
y = np.array(([92], [86], [89]), dtype=float) / 100

sigmoid = lambda x: 1/ (1+ np.exp(-x))
sigmoid_der = lambda x: x/ (1 - x)

# define epochs and learning rate
epochs, learning_rate = 5, 0.1

# neurons count
input_neurons, hidden_neurons, output_neurons = 2,3,1

# define weights and bias
ihw, hb = np.random.uniform(size=(input_neurons, hidden_neurons)), np.random.uniform(size=(1, hidden_neurons))
how, ob = np.random.uniform(size=(hidden_neurons, output_neurons)), np.random.uniform(size=(1, output_neurons))

# training loop
for epoch in range(epochs):
    hlayer_act = sigmoid((X @ ihw) + hb)
    output = sigmoid((hlayer_act @ how) + ob)

    err_out = y - output
    d_output = err_out * sigmoid_der(output)
    err_hid = (d_output @ how.T) * sigmoid_der(hlayer_act)

    ihw += (X.T @ err_hid) * learning_rate
    how += (hlayer_act.T @ d_output) * learning_rate

    print(f"Epoch {epoch+1} starts ---")
    print(f"Input: {X}")
    print(f"Actual output: {y}")
    print(f"Predicted output: {output}")
    print(f"Epoch {epoch+1} ends ----")
```
